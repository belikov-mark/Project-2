```python
#Question 2 Who had the highest 3 point percentage on The Milwakuee Bucks during their 2020-2021 Chammpionship year
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import numpy as np

#Sends a request to the webpage
url = "https://www.basketball-reference.com/teams/MIL/2021.html"
response = requests.get(url)

#Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

#Finds the id that gets the stats
table = soup.find("table", id="per_game")

if table is not None:
#Finds all rows in the table body
    rows = table.find("tbody").find_all("tr")

#lists to store player names and 3 Point Percentage
    player_names = []
    three_point_percentages = []

    for row in rows:
#Extracts data from each cell in the row
        cols = row.find_all("td")
        
#Checks if the row contains player statistics
        if len(cols) > 0:
            player_name = cols[0].text.strip()
            three_point_percentage_index = None
            for i, col in enumerate(cols):
                if col.get("data-stat") == "fg3_pct":
                    three_point_percentage_index = i
                    break

            if three_point_percentage_index is not None:
                three_point_percentage = float(cols[three_point_percentage_index].text.strip())
                player_names.append(player_name)
                three_point_percentages.append(three_point_percentage)

#Create a colormap with different shades of green
    cmap = plt.get_cmap('Greens')
    colors = cmap(np.linspace(0, 1, len(player_names)))

#Creates a bar plot for all of the players
    plt.figure(figsize=(12, 8))
    plt.bar(player_names, three_point_percentages, color=colors)
    plt.title("Three-Point Shooting Percentage for Milwaukee Bucks Players (2020-2021)")
    plt.xlabel("Player")
    plt.ylabel("Three-Point Shooting Percentage")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
else:
    print("Table not found with id 'per_game'.")
```
