Take data from basketball reference to answer questions.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

#Sends request to website
url = "https://www.basketball-reference.com/leagues/NBA_stats_per_game"
response = requests.get(url)

#Takes the content and makes it so that you can use it in your program
soup = BeautifulSoup(response.content, "html.parser")

#Gets the stats
table = soup.find("table", id="stats")

#Gets the column for the question
colgroup = table.find("<colgroup>")
col_elements = colgroup.find_all("<col>")

#3PTA per game column
three_point_attempts_index = None
for i in range(len(col_elements)):
    col = col_elements[i]
    if col.get("data-stat") == "fg3a_per_g":
        three_point_attempts_index = i
        break

if three_point_attempts_index is not None:
#Takes data for each NBA season
    data = []
#Find all rows in the table body
    rows = table.find("tbody").find_all("tr")
    for row in rows:
#Take data from each cell in the row
        cols = row.find_all("td")
        season = cols[0].text.strip()
        three_point_attempts_per_game = cols[three_point_attempts_index].text.strip()
        data.append({"Season": season, "3PA Per Game": three_point_attempts_per_game})

#Create a DataFrame from the extracted data
    df = pd.DataFrame(data)
#Display the DataFrame
    print(df.head())
```
